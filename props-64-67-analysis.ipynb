Filtering the propositions to prop 67

```python
top_supporters64 = support.groupby(["contributor_firstname", "contributor_lastname"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
top_supporters64.head(10)
```

```python
top_supporters64.amount.plot.bar()
```

```python
top_supporters64.amount.plot.barh()
```

```python
top_supporters64.head(7).amount.plot.barh()
```

```python
chart64 = top_supporters64.head(7).amount.plot.barh()
chart64.set_yticklabels(top_supporters64.contributor_firstname)
```

```python
top_supporters64['contributor_fullname'] = top_supporters64.contributor_firstname + " " + top_supporters64.contributor_lastname
```

```python
top_supporters64.head(10)
```

```python
chart = top_supporters64.head(7).amount.plot.barh()
chart.set_yticklabels(top_supporters64.contributor_fullname)
```

```python
def combine_names(row):
    if row.contributor_fullname.startswith('SEAN PARKER'): return 'SEAN PARKER' 
    return row.contributor_fullname
```

```python
top_supporters64.apply(combine_names, axis=1)
```

```python
top_supporters64['contributor_cleanname'] = top_supporters64.apply(combine_names, axis=1)
```

```python
top_supporters64.groupby('contributor_cleanname').amount.sum().reset_index().sort_values("amount", ascending=False).head(7)
```
